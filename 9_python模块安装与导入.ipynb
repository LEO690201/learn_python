##### 库
- pip是python的包管理工具，可以帮助我们安装、卸载、管理python第三方库。
```bash
pip --version # 查看pip版本
pip install <requests> # 安装requests模块
pip uninstall <requests> # 卸载requests模块
pip list # 查看已安装的模块
```
##### 实用库
