```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,10)    #生成100个0到10之间的分成100段的数,同matlab的linspace函数
print(x)
y=np.sin(x)
plt.plot(x,y)
plt.title("sin(x)")       #设置标题
plt.xlabel("x")           #设置x轴标签
plt.ylabel("sin(x)")      #设置y轴标签
plt.grid(True)            #显示网格
plt.show()
```

```python
#
plt.scatter(x, y)    # 绘制散点图
plt.show()
plt.scatter(x, y, marker='*')
plt.show()
plt.scatter(x, y, marker='*', s=100)    # s参数控制点的大小
plt.show()
plt.scatter(x, y, marker='*', s=100, c='r')    # c参数控制点的颜色
plt.show()
plt.scatter(x, y, marker='*', s=100, c='r', alpha=0.5)    # alpha参数控制点的透明度
plt.show()
plt.scatter(x, y, marker='*', s=100, c='r', alpha=0.5, label='scatter')    # label参数为图例
plt.legend()    # 显示图例
plt.show()
plt.scatter(x, y, marker='*', s=100, c='r', alpha=0.5, label='scatter')
plt.plot(x, y, label='line')    # 绘制折线图
plt.legend()
plt.show()
plt.scatter(x, y, marker='*', s=100, c='r', alpha=0.5, label='scatter')
plt.plot(x, y, label='line')
plt.legend(loc='upper left')    # 图例位置
plt.show()
plt.scatter(x, y, marker='*', s=100, c='r', alpha=0.5, label='scatter')
plt.plot(x, y, label='line')
plt.legend(loc='upper left', fontsize=12)    # 图例字体大小
plt.show()
plt.scatter(x, y, marker='*', s=100, c='r', alpha=0.5, label='scatter')
plt.plot(x, y, label='line')
plt.legend(loc='upper left', fontsize=12, ncol=2)    # 图例列数
plt.show()
plt.scatter(x, y, marker='*', s=100, c='r', alpha=0.5, label='scatter')
plt.plot(x, y, label='line')
plt.legend(loc='upper left', fontsize=12, ncol=2, shadow=True)    # 阴影
plt.show()
plt.scatter(x, y, marker='*', s=100, c='r', alpha=0.5, label='scatter')
plt.plot(x, y, label='line')
plt.legend(loc='upper left', fontsize=12, ncol=2, shadow=True, fancybox=True)    # 边框
plt.show()
plt.scatter(x, y, marker='*', s=100, c='r', alpha=0.5, label='scatter',linestyle='--')    # 线型
plt.plot(x, y, label='line',linestyle='--')
plt.legend(loc='upper left', fontsize=12, ncol=2, shadow=True, fancybox=True)
plt.show()
```

```python
x=np.linspace(0,10,10)
y=np.sin(x)

x2=np.linspace(0,10,100)
y2=np.sin(x2)

plt.plot(x,y)
plt.plot(x2,y2)
```

```python
fig,axes=plt.subplots(1,2)     # 创建一个包含1行2列的子图
axes[0].plot(x,y)             # 绘制第一个子图
axes[1].plot(x2,y2)             # 绘制第二个子图
plt.show()                     # 显示图形
```
